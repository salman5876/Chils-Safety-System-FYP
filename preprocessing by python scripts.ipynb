# Rename all images in folder

```python
# This code can rename all images in a folder 

import os

def rename_images(folder_path, new_name):
    # Get a list of all files in the folder
    files = os.listdir(folder_path)
    
    # Filter out non-image files
    image_files = [file for file in files if file.endswith(('.jpg', '.jpeg', '.png', '.gif'))]
    
    # Iterate through the image files and rename them
    for i, file in enumerate(image_files, start=1):
        # Extract the file extension
        file_extension = os.path.splitext(file)[1]
        
        # Generate the new file name
        new_file_name = f"{new_name}{i}{file_extension}"
        
        # Build the old and new file paths
        old_file_path = os.path.join(folder_path, file)
        new_file_path = os.path.join(folder_path, new_file_name)
        
        # Rename the file
        os.rename(old_file_path, new_file_path)
        
        print(f"Renamed '{file}' to '{new_file_name}'")

# Get the folder path and new name from the user
folder_path = input("Enter the folder path: ")
new_name = input("Enter the new name for the images: ")

# Call the rename_images function
rename_images(folder_path, new_name)

```

# Convert images extention to jpg of whole folder

```python
#by this code we can convert hundreds of images to jpg extention 

from PIL import Image
import os

# Folder containing the images to be converted
input_folder = r"C:\Users\Salman Ahmed\Desktop\vid img"

# Output folder to save the converted JPG images
output_folder = r"C:\Users\Salman Ahmed\Desktop\vid img1"

# Ensure the output folder exists, create it if not
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Supported image extensions
supported_extensions = [".webp", ".avif", ".jpg", ".jpeg", ".png"]

# Loop through files in the input folder
for filename in os.listdir(input_folder):
    file_extension = os.path.splitext(filename)[1].lower()
    
    # Check if the file has a supported image extension
    if file_extension in supported_extensions:
        try:
            # Open the image and convert it to RGB mode (if not already)
            image = Image.open(os.path.join(input_folder, filename))
            if image.mode != 'RGB':
                image = image.convert('RGB')
            
            # Save the image as a JPG file in the output folder
            jpg_filename = os.path.splitext(filename)[0] + ".jpg"
            image.save(os.path.join(output_folder, jpg_filename), "JPEG")
            print(f"Converted {filename} to {jpg_filename}")
        except Exception as e:
            print(f"Error converting {filename}: {str(e)}")
    
print("Conversion complete.")
```

#  Resize multiple images at once

```python
# This code can resize multiple image to the given height and width 

from PIL import Image
import os

# Set the target size
target_width = 640
target_height = 640

# Input and output directories
input_folder = r"C:\Users\Salman Ahmed\Desktop\datasets\baby items\dataset\images"  # Change this to your input folder
output_folder = r"C:\Users\Salman Ahmed\Desktop\datasets\baby items\dataset"  # Change this to your output folder
    

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop through all files in the input folder
for filename in os.listdir(input_folder):
    # Check if the file is an image (you can add more file extensions if needed)
    if filename.endswith((".jpg", ".jpeg", ".png", ".bmp")):
        # Open the image using Pillow
        with Image.open(os.path.join(input_folder, filename)) as img:
            # Resize the image to the target size
            img = img.resize((target_width, target_height))
            # Save the resized image to the output folder
            img.save(os.path.join(output_folder, filename))

print("Image resizing completed.")
```

# Filters for images enhancement

```python
# Gaussian Blur filter

import cv2
import os

# Define the folder containing your images
folder_path = r"C:\Users\Salman Ahmed\Desktop\sfsd"

# Define the output folder where blurred images will be saved
output_folder = r"C:\Users\Salman Ahmed\Desktop\sfsd"

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Define the Gaussian blur kernel size
kernel_size = (5, 5)  # Adjust this size according to your needs

# Iterate over each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif')):
        # Load the image
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path)

        # Apply Gaussian Blur
        blurred_image = cv2.GaussianBlur(image, kernel_size, 0)

        # Define the output path for the blurred image
        output_path = os.path.join(output_folder, filename)

        # Save the blurred image
        cv2.imwrite(output_path, blurred_image)

        print(f'Applied Gaussian Blur to {filename} and saved as {output_path}')

print('All images processed.')
```

```python
# Contrast Enhancement Filters:
import cv2
import os
import numpy as np

# Define the input folder containing your images
input_folder = r"C:\Users\Salman Ahmed\Desktop\main dataset\images\train  first"

# Define the output folder where enhanced images will be saved
output_folder = r"C:\Users\Salman Ahmed\Desktop\main dataset\images\val first"

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Loop through all files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif')):
        # Load the image
        image_path = os.path.join(input_folder, filename)
        image = cv2.imread(image_path)

        # Convert the image to grayscale (if it's not already)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Apply contrast stretching
        min_intensity = 0
        max_intensity = 255
        min_original = np.min(gray)
        max_original = np.max(gray)

        # Calculate the scaling factors for contrast stretching
        alpha = (max_intensity - min_intensity) / (max_original - min_original)
        beta = min_intensity - alpha * min_original

        # Apply contrast stretching
        contrast_stretched = cv2.convertScaleAbs(gray, alpha=alpha, beta=beta)

        # Convert the contrast-stretched image back to BGR format (if needed)
        contrast_stretched_bgr = cv2.cvtColor(contrast_stretched, cv2.COLOR_GRAY2BGR)

        # Define the output path for the enhanced image
        output_path = os.path.join(output_folder, filename)

        # Save the enhanced image
        cv2.imwrite(output_path, contrast_stretched_bgr)

        print(f'Applied Contrast Enhancement to {filename} and saved as {output_path}')

print('All images processed.')
```

# Copy only jpg image form folder and paste to another folder 

```python
Basically this code can seprate jpg images from other extention images, it check a whole folder and paste jpg images to another folder
import os
import shutil

# Source folder containing the JPG images
source_folder = r"C:\Users\Salman Ahmed\Desktop\datasets\furir\test_zip\test"

# Destination folder where you want to copy the JPG images
destination_folder = r"C:\Users\Salman Ahmed\Desktop\main dataset\images\val first"

# Ensure the destination folder exists, or create it if necessary
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# Loop through files in the source folder
for filename in os.listdir(source_folder):
    # Check if the file is a JPG image (case insensitive)
    if filename.lower().endswith('.jpg'):
        # Build the full path for the source and destination files
        source_file = os.path.join(source_folder, filename)
        destination_file = os.path.join(destination_folder, filename)
        
        # Copy the file to the destination folder
        shutil.copy(source_file, destination_file)
        print(f'Copied: {filename} to {destination_folder}')
```

# Check which images have not corresponding annotation files

```python
# when we are working with multiples images annotation we faced problem of misbalance images 
# so this code help to detect which image is annoted or which is not. 

import os

# Path to the folder containing images
image_folder = r"C:\Users\Salman Ahmed\Desktop\main dataset\images\train"

# Path to the folder containing annotations
annotation_folder = "C:\Users\Salman Ahmed\Desktop\main dataset\labels\train"

# Get a list of all image files in the image folder
image_files = [f for f in os.listdir(image_folder) if f.endswith('.jpg')]

# Get a list of all annotation files in the annotation folder
annotation_files = [f for f in os.listdir(annotation_folder) if f.endswith('.txt')]

# Extract the base names of image files (without extensions)
image_base_names = [os.path.splitext(os.path.basename(f))[0] for f in image_files]

# Extract the base names of annotation files (without extensions)
annotation_base_names = [os.path.splitext(os.path.basename(f))[0] for f in annotation_files]

# Find image files without corresponding annotation files
missing_annotations = [img for img in image_base_names if img not in annotation_base_names]

# Print the list of image files without annotations
for img_name in missing_annotations:
    print(f"Image {img_name}.jpg does not have a corresponding annotation.")

# Optionally, you can also save the list to a text file
# with open('missing_annotations.txt', 'w') as file:
#     for img_name in missing_annotations:
#         file.write(f"Image {img_name}.jpg does not have a corresponding annotation.\n")
```

# Check and delete images having not corresponding files

```python
# when we are working with multiples images annotation we faced problem of misbalance images 
# so this code help to detect which image is not annotated and delete that image. 

import os

# Path to the folder containing images
image_folder = r"C:\Users\Salman Ahmed\Desktop\main dataset\images\train"

# Path to the folder containing annotations
annotation_folder = r"C:\Users\Salman Ahmed\Desktop\main dataset\labels\train"

# Get a list of all image files in the image folder
image_files = [f for f in os.listdir(image_folder) if f.endswith('.jpg')]

# Get a list of all annotation files in the annotation folder
annotation_files = [f for f in os.listdir(annotation_folder) if f.endswith('.txt')]

# Extract the base names of image files (without extensions)
image_base_names = [os.path.splitext(os.path.basename(f))[0] for f in image_files]

# Extract the base names of annotation files (without extensions)
annotation_base_names = [os.path.splitext(os.path.basename(f))[0] for f in annotation_files]

# Find image files without corresponding annotation files
missing_annotations = [img for img in image_base_names if img not in annotation_base_names]

# Print the list of image files without annotations
for img_name in missing_annotations:
    print(f"Image {img_name}.jpg does not have a corresponding annotation.")
    
# Delete image files without corresponding annotations
for img_name in missing_annotations:
    img_path = os.path.join(image_folder, img_name + '.jpg')
    if os.path.exists(img_path):
        os.remove(img_path)
        print(f"Deleted image {img_name}.jpg.")

# Optionally, you can also save the list to a text file
# with open('missing_annotations.txt', 'w') as file:
#     for img_name in missing_annotations:
#         file.write(f"Image {img_name}.jpg does not have a corresponding annotation.\n")
```
